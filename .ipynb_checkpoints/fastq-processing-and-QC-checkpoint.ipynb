### Plot complexity curves

```r
#load R packages
library(tidyverse)
library(ggsci)
```

```r
#read function
read_lcExtrap <- function(input_file, condition, library_type){
  #This function reads in the lcextrap txt file and assigns descriptors of the data to new columns for use in plotting. 
  #condition example is "GM12878inGM12878", library_type example is "DNA_Rep1"
  return(read_tsv(input_file) %>% mutate(condition=condition, Type=library_type))
}
```

```r
#GG only for the method paper. 

#Read in results and assign to unique data.frame objects. 
setwd("/data/hodges_lab/ATAC-STARR_B-cells/results/library_complexity")
GGD1 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_DNA_Rep1_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "DNA Rep1")
GGD2 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_DNA_Rep2_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "DNA Rep2")
GGD3 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_DNA_Rep3_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "DNA Rep3")
GGR1 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_RNA_Rep1_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "RNA Rep1")
GGR2 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_RNA_Rep2_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "RNA Rep2")
GGR3 <- read_lcExtrap(input_file = 'lc-extrap/GM12878inGM12878_RNA_Rep3_lc-extrap-results.txt', condition = "GM12878inGM12878", library_type = "RNA Rep3")

#Merge into one dataframe. 
lcExtrap_all <- bind_rows(GGD1, GGD2, GGD3, GGR1, GGR2, GGR3)

#Plot:
lcExtrap_plot <- ggplot(lcExtrap_all) + 
  geom_ribbon(aes(x=TOTAL_READS/10^6, ymin=LOWER_0.95CI/10^6, ymax=UPPER_0.95CI/10^6, fill=Type)) +
  geom_line(aes(x=TOTAL_READS/10^6, y=EXPECTED_DISTINCT/10^6, colour=Type), size=1, linetype="dashed") +
  xlab("Sequenced reads (M)") + 
  ylab("Distinct reads (M)") +
  ggtitle("Estimated Complexity of ATAC-STARR-seq Libraries") +
  theme_bw(base_size = 6, base_family = "Arial") +
  theme(plot.title = element_text(hjust = 0.5, size = 12, face = "bold")) +
  coord_cartesian(xlim = c(0, 1000), ylim = c(0,100)) +
  scale_color_npg() +
  scale_fill_npg(alpha = 0.25)

lcExtrap_plot
```

```r
#save plot
ggsave(filename = "/data/hodges_lab/ATAC-STARR_B-cells/results/library_complexity/GG-AS-samples_complexity-curve.pdf", 
    plot = lcExtrap_plot, units = "in", width = 4, height = 3, device = cairo_pdf)
```

### Plot insert size distribution

```r
#load R packages
library(tidyverse)
library(ggsci)
```

```r
#read function
read_ISM <- function(input_file, condition, library_type){
  #This function reads in the lcextrap txt file and assigns descriptors of the data to new columns for use in plotting. 
  #condition example is "GM12878inGM12878", library_type example is "DNA_Rep1"
  tsv <- read_tsv(input_file, skip = 12, col_names = FALSE) %>% mutate(condition=condition, Type=library_type)
  return(mutate(tsv, cpm=(X2/sum(tsv$X2))*1e6))
}
```

```r
#GG only for the method paper. 

#Read in results and assign to unique data.frame objects. 
setwd("/data/hodges_lab/ATAC-STARR_B-cells/results/library_complexity/")
GGD1 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep1_picard-CISM.txt', condition = "DNA", library_type = "Rep1")
GGD2 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep2_picard-CISM.txt', condition = "DNA", library_type = "Rep2")
GGD3 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep3_picard-CISM.txt', condition = "DNA", library_type = "Rep3")
GGR1 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep1_picard-CISM.txt', condition = "RNA", library_type = "Rep1")
GGR2 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep2_picard-CISM.txt', condition = "RNA", library_type = "Rep2")
GGR3 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep3_picard-CISM.txt', condition = "RNA", library_type = "Rep3")

#Merge into one dataframe. 
ISM_all <- bind_rows(GGD1, GGD2, GGD3, GGR1, GGR2, GGR3)

#Plot:
ISM_plot <- ggplot(ISM_all, aes(x=X1, y=cpm)) +
  geom_line(size = 0.5) +
  facet_grid(rows = vars(Type), cols = vars(condition)) +
  coord_cartesian(x=c(0,500)) +
  scale_x_continuous(breaks = seq(0, 500, by = 100)) +
  scale_color_manual(values = c("black", "black", "black", "black", "black", "black")) +
  xlab("Insert Size (bp)") + 
  ylab("Counts per Million") +
  ggtitle("Insert Size Distribution of ATAC-STARR-seq Libraries") +
  theme_bw(base_size = 8, base_family = "Arial") +
  theme(plot.title = element_text(hjust = 0.5, size = 12, face = "bold"), 
        strip.text.x = element_text(size = 8, family = "Arial", face = "italic", color = "black"),
        strip.text.y = element_text(size = 8, family = "Arial", face = "plain", color = "black"),
        strip.background = element_rect(fill = "lightgray"), 
        legend.position = "none")

ISM_plot
```

```r
ggsave(filename = "/data/hodges_lab/ATAC-STARR_B-cells/results/library_complexity/InsertSizeMetrics_gg-samples.pdf", 
    plot = ISM_plot, units = "in", width = 4, height = 3, device = cairo_pdf)
```

```r
#Read in results and assign to unique data.frame objects. 
setwd("/data/hodges_lab/ATAC-STARR_B-cells/results/library_complexity")
GGD1 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep1_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "DNA Rep1")
GGD2 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep2_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "DNA Rep2")
GGD3 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_DNA_Rep3_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "DNA Rep3")

GGR1 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep1_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "RNA Rep1")
GGR2 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep2_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "RNA Rep2")
GGR3 <- read_ISM(input_file = 'picard-CISM/GM12878inGM12878_RNA_Rep3_picard-CISM.txt', condition = "GM12878inGM12878", library_type = "RNA Rep3")

GLD1 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_DNA_Rep1_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "DNA Rep1")
GLD2 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_DNA_Rep2_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "DNA Rep2")
GLD3 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_DNA_Rep3_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "DNA Rep3")

GLR1 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_RNA_Rep1_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "RNA Rep1")
GLR2 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_RNA_Rep2_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "RNA Rep2")
GLR3 <- read_ISM(input_file = 'picard-CISM/GM12878inLCL8664_RNA_Rep3_picard-CISM.txt', condition = "GM12878inLCL8664", library_type = "RNA Rep3")

LGD1 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_DNA_Rep1_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "DNA Rep1")
LGD2 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_DNA_Rep2_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "DNA Rep2")
LGD3 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_DNA_Rep3_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "DNA Rep3")

LGR1 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_RNA_Rep1_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "RNA Rep1")
LGR2 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_RNA_Rep2_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "RNA Rep2")
LGR3 <- read_ISM(input_file = 'picard-CISM/LCL8664inGM12878_RNA_Rep3_picard-CISM.txt', condition = "LCL8664inGM12878", library_type = "RNA Rep3")

LLD1 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_DNA_Rep1_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "DNA Rep1")
LLD2 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_DNA_Rep2_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "DNA Rep2")
LLD3 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_DNA_Rep3_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "DNA Rep3")

LLR1 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_RNA_Rep1_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "RNA Rep1")
LLR2 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_RNA_Rep2_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "RNA Rep2")
LLR3 <- read_ISM(input_file = 'picard-CISM/LCL8664inLCL8664_RNA_Rep3_picard-CISM.txt', condition = "LCL8664inLCL8664", library_type = "RNA Rep3")

#Merge into one dataframe. 
ISM_all <- bind_rows(GGD1, GGD2, GGD3, GGR1, GGR2, GGR3, GLD1, GLD2, GLD3, GLR1, GLR2, GLR3, 
    LGD1, LGD2, LGD3, LGR1, LGR2, LGR3, LLD1, LLD2, LLD3, LLR1, LLR2, LLR3)

#Plot:
ISM_plot <- ggplot(ISM_all, aes(x=X1, y=cpm)) +
  geom_line(aes(colour=Type), size = 0.5) +
  facet_grid(rows = vars(Type), cols = vars(condition)) +
  coord_cartesian(x=c(0,500)) +
  scale_x_continuous(breaks = seq(0, 500, by = 100)) +
  scale_color_manual(values = c("black", "black", "black", "black", "black", "black")) +
  xlab("Insert Size (bp)") + 
  ylab("Counts per Million") +
  ggtitle("Insert Size Distribution of ATAC-STARR-seq Libraries") +
  theme_bw(base_size = 8, base_family = "Arial") +
  theme(plot.title = element_text(hjust = 0.5, size = 12, face = "bold"), 
        strip.text.x = element_text(size = 8, family = "Arial", face = "italic", color = "black"),
        strip.text.y = element_text(size = 8, family = "Arial", face = "plain", color = "black"),
        strip.background = element_rect(fill = "lightgray"), 
        legend.position = "none")
        
ISM_plot
```
